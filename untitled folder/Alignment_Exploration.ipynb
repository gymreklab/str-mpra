## Helper Functions

````python
def ReverseComplement(sequence):
    basecomplement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'} 
    complement = ''.join([basecomplement[base] for base in sequence])
    revcomp = complement[::-1]
    
    return revcomp
````

````python
def BC_check (in_df):

    STR_BCs = []
    BCs = []
    STRs = []
    occurence = []

    for STR, barcode in zip(in_df.STR, in_df.barcode):
        STR_BC = tuple([STR, barcode])

        if STR_BC not in STR_BCs:
            STR_BCs.append(STR_BC)
            BCs.append(barcode)
            STRs.append(STR)
            occurence.append(1)
        else:
            occurence[STR_BCs.index(STR_BC)] += 1

    check = pd.DataFrame({"STR_BC": STR_BCs,
                          "barcode": BCs, 
                          "STR": STRs,
                          "occurence": occurence})
    
    return check
````

````python
def remove_dup_BC (check):
    duplicate_barcode = check[check.duplicated('barcode')]
    remove_dup = set(duplicate_barcode.barcode)
    association_df = check[~check["barcode"].isin(remove_dup)]
    
    return association_df
````

````python
def STR_BC_dict (association_df):
    association = {}

    for STR_BC in association_df.STR_BC:
        barcode = STR_BC[1]
        STR = STR_BC[0]

        association[barcode] = STR

    print(str(len(association)) + " unique barcode is matched to " + 
          str(len(set(association.values()))) + " unique STR")
    
    return association
````

````python
def check_overlap (data, length):
    overlap = []
    read1s = list(data["R1"].tolist())
    read2s = list(data["R2"].tolist())
    
    for read1, read2 in zip(read1s, read2s):
        if read1[0:int(length)] == read2[-int(length):]:
            overlap.append(True)
        else:
            overlap.append(False)
    
    data["overlap"] = overlap
````

````python
def countplot_bc_association (association_df):
    plot_df = association_df.STR.value_counts().to_frame().reset_index()
    plot_df.columns = ["STR", "count"]
    ax = sns.countplot(data=plot_df, x="count")
    plt.xlabel("# barcode associate with a unique STR");
    plt.ylabel("count of STR with\ncertain number of associated barcode");  
    plt.title("How many STR have multiple barcode?")
    
    ax2 = plt.axes([0.3, 0.3, 0.55, 0.55])
    sns.countplot(data=plot_df, x="count", ax = ax2)
    ax2.set_title('zoom')
    ax2.set_xlabel(None)
    ax2.set_ylabel(None)
    ax2.set_ylim([0,120]);
````

````python
def countplot_STRBC_occurence (association_df):
    ax = sns.countplot(data=association_df, x="occurence");
    plt.title("How many STR-BC paired occurred multiple time?")
    
    ax2 = plt.axes([0.3, 0.3, 0.55, 0.55])
    sns.countplot(data=association_df, x="occurence", ax = ax2)
    ax2.set_title('zoom')
    ax2.set_xlabel(None)
    ax2.set_ylabel(None)
    ax2.set_ylim([0,120]);
    plt.axvline(10, color="black", linestyle="--")
````

## Filtering alignment score: 
- defined in Abell et al as: __low-quality alignment scores (<100)__
    >- which for STAR, the mapping quality in column 5 is 255 for uniquely mapping reads
- for bwamem, the mapping quality in column 5 is phred scale, which according to [Board Institute](https://gatk.broadinstitute.org/hc/en-us/articles/360035531872-Phred-scaled-quality-scores)
    >- "a Phred Score of 20 or above is acceptable, because this means that whatever it qualifies is 99% accurate, with a 1% chance of error" 
    >- Hence, __filter those with MapQ < 20__
- for bowtie2
    >- “Bowtie2 does not use the number of times a read mapped in the calculation for MAPQ. Instead, it just uses the AS and the XS, which is either >= the minimum score or below it. If AS == XS, the read is considered a true multiread and can only get a score of 0 or 1. If XS is below the minimum score, the read is considered a true uniread under the given scoring scheme. True unireads can get scores of 0, 3, 8, 23, 24, 40 and 42. A score of 0 is reserved for the true unireads with an AS in the bottom 30% of allowable scores. The scores 3, 8, 23, 24, 40, and 42 are unique to true unireads.” [Reference](http://biofinysics.blogspot.com/2014/05/how-does-bowtie2-assign-mapq-scores.html)
    >- So, kept those with mapq of 0 and 3 (?)
