## Association & Analysis

````python
merge = pd.merge(left=R2, right=R1,
                 how='left', left_on='read_id',
                 right_on='read_id')
merge
````

````python
# see if STR does not have a barcode 
# which is given by the barcode count 
merge.isnull().sum()
````

````python
# check association
temp = merge.dropna()

STR_BCs = []
STRs = []
BCs = []
occurence = []

for STR, barcode in zip(temp.STR, temp.barcode):
    STR_BC = tuple([STR, barcode])
        
    if STR_BC not in STR_BCs:
        STR_BCs.append(STR_BC)
        STRs.append(STR)
        BCs.append(barcode)
        occurence.append(1)
    else:
        occurence[STR_BCs.index(STR_BC)] += 1

check = pd.DataFrame({"STR_BC": STR_BCs,
                      "barcode": BCs, 
                      "STR": STRs,
                      "occurence": occurence})
display(temp)
display(check)
````

### Association dictionary with barcode as key and STR name as value
>- duplicates are removed

````python
remove_dup = set(duplicate_barcode.barcode)
association_df = check[~check["barcode"].isin(remove_dup)]
association_df.barcode.is_unique
````

````python
association = {}

for STR_BC in association_df.STR_BC:
    barcode = STR_BC[1]
    STR = STR_BC[0]
    
    association[barcode] = STR

print(str(len(association)) + " unique barcode is matched to a STR")
print(str(len(set(association.values()))) + " unique STR")
````
